# Matplotlib


**Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+**

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,10)
y = x**2
```

```python
x
```

```python
y
```

```python
#fucntional way to create a plot
plt.plot(x,y)
```

```python
#color as 3rd argument
```

```python
plt.plot(x,y,'r')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'G')
```

```python
#oops method
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.5])
axes.plot(x,y)
axes.set_xlabel("x_label")
axes.set_ylabel("y_label")
axes.set_title("Graph")
```

```python
#multiple plots
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.9,0.9])
axes2 = fig.add_axes([0.2,0.6,0.3,0.3])

axes1.plot(x,y)
axes1.set_title('Outer Plot')

axes2.plot(x,y)
axes2.set_title('Inner Plot')
```

```python
#subplots
```

```python
fig,axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_title("First subplot")
axes[1].plot(y,x)
axes[1].set_title("Second subplot")

plt.tight_layout() #takes care of overlapping subplots
```

```python
axes # since axes is array you can use for loop to traverse through plots
```

# FigureSize and DPI

```python
#fig = plt.figure(figsize = (7,3)) #works with subplots too
fig,axes = plt.subplots(nrows = 2,ncols = 1,figsize = (6,3))
#ax = fig.add_axes([0,0,1,1])
axes[0].plot(x,y)
axes[1].plot(x,y)

plt.tight_layout()
```

```python
#legends
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,x**2,label='X squared')
axes.plot(x,x**3,label='X cubed')
axes.set_xlabel("x_label")
axes.set_ylabel("y_label")
axes.legend()
axes.set_title("Graph")
```

```python
#Matplotlib Design features
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.9,0.9])

axes.plot(x,y,color="purple",lw=2,ls="--",marker='o',markersize=15,markerfacecolor='green') 
# lw = line width
# ls= line spacing
```
